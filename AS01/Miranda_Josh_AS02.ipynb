````python
import math
import time

def naive(n):
    start_time = time.time()
    #print("Start:", start_time)
    
    s = 0
    print("N:", str(n))
    
    for i in range(0,n):
        s = s + math.sqrt(19.0)
        
    print("Summation:", s)
    
    end_time = time.time()
    #print("End:", end_time)
    
    elapsed_time = end_time - start_time
    print("Elapsed Time:", elapsed_time, "seconds")
    
    exact_sum = n * math.sqrt(19.0)
    print("Exact:", exact_sum)
       
    absolute_error = abs(s-exact_sum)
    print("Absolute Error:", absolute_error)
        
    relative_error = absolute_error/exact_sum
    print("Relative Error:", relative_error)
    
    print("\n")
    
def comp(n):
    start_time = time.time()
    #print("Start:", start_time)
    
    s = 0
    print("N:", str(n))
    
    e = 0
    
    for i in range(0,n):
        temp = s
        y = math.sqrt(19.0) + e
        s = temp + y
        e = (temp - s) + y
    s += e
    print("Summation:", s)
    
    end_time = time.time()
    #print("End:", end_time)
    
    elapsed_time = end_time - start_time
    print("Elapsed Time:", elapsed_time, "seconds")
    
    exact_sum = n * math.sqrt(19.0)
    print("Exact:", exact_sum)
    
    absolute_error = abs(s-exact_sum)
    print("Absolute Error:", absolute_error)
    
    relative_error = absolute_error/exact_sum
    print("Relative Error:", relative_error) 
    
    print("\n")
    
    
def both():    
    N = [10**3,10**4,10**5,10**6,10**7,10**8]
    for x in N:
        print("Naive Algorithm")
        naive(x)
        print("Compensated Algorithm")
        comp(x)
    

print(both())
    


    
````

In order to test these methods the summations with be perform on N = $10^{3}$, $10^{4}$, $10^{5}$, $10^{6}$, $10^{7}$, and $10^{8}$.



**Naive Summation**

```Python
s = 0
print("N:", str(n))
    
for i in range(0,n):
    s = s + math.sqrt(19.0)
        
print("Summation:", s)
```

Here is the naive algorithm where it will keep adding $\sqrt{19}$ for range 0 to n.



**Compensated Summation**

```Python
s = 0
print("N:", str(n))
e = 0
    
for i in range(0,n):
    temp = s
    y = math.sqrt(19.0) + e
    s = temp + y
    e = (temp - s) + y
s += e
print("Summation:", s)
```
    
The algorithm for compensated summation which is recursive. Similar to the naive algorithm it will add $\sqrt{19}$ for range 0 to n but this time there is a correction term designed to diminish rounding errors, seen in the variable e.
